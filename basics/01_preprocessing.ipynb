```python
import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col]= df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['bought'])
le.transform(df['bought'])
```

```python
le.fit_transform(df['bought'])
```

```python
le.classes_
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

Powrt do poprzedniej postaci

```python
le.inverse_transform(df['bought'])
```

```python
df['bought'] = le.inverse_transform(df['bought'])
df
```

Label encoder jest uzywana do przeksztacania gwnie zmiennych docelowych - w tym przypadku bought. Do reszty mamy inne biblioteki.


**Kodowanie zmiennych kategorycznycy OneHotEncoder**

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
#jeli sparse=True otrzymujemy macierz rzadk - macierz, w ktrej wikszo elementw ma warto zero.
encoder.fit(df[['size']])
```

```python
encoder.transform(df[['size']])
```

```python
encoder.categories_
```

```python
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoder.fit_transform(df[['size']])
```

**Standaryzacja - StandardScaler**

```python
print(f"{df['price']}\n")
print(f"rednia: {df['price'].mean()}")
print(f"Odchylenie: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean())/df['price'].std()
```

```python
def standarize(value):
  return (value - value.mean())/value.std()

standarize(df['price'])
```

```python
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit_transform(df[['price']])
```

```python
scaler_price = StandardScaler()
df['price'] = scaler_price.fit_transform(df[['price']])
scaler_weight = StandardScaler()
df['weight'] = scaler_weight.fit_transform(df[['weight']])
df
```

Wykonanie standaryzacji dla 2 kategorii na raz

```python
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df
```
